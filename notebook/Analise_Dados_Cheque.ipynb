```python
import os
import pandas as pd
```

```python
import os.path
def path_base(base_name):
    current_dir = os.path.abspath(os.path.join(os.getcwd()))
    print(current_dir)
    data_dir = current_dir.replace('notebook','data')
    print(data_dir)
    data_base = data_dir + '\\' + base_name
    print(data_base)
    return data_base
```

```python
stats = pd.read_csv(path_base('db_analise_dados_cheque.csv'),converters={"cheque":float})
```
